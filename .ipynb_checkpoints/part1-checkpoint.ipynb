```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))

def logistic_loss(X ,y , W):
    pred = sigmoid(X @ W)
    ones = np.ones((y.shape[0],1))
    loss =(-((y.T@np.log(pred)) + (ones-y).T@np.log(ones - pred))/(y.shape[0]))
    return loss
```

```python
def gradient_descent(X, y, W, learning_rate=0.1, num_epochs=10 , verbose=0):
    m = X.shape[0]
    losses = []
    for _ in range(num_epochs):
        pred = sigmoid(X @ W)
        grad = (1/m)*(X.T @(pred - y))
        W = W - (learning_rate)*grad
        losses.append(logistic_loss(X, y, W))
        if verbose!=0:
            print(logistic_loss(X, y, W)[0][0],end="->")
    return losses ,W
```

```python
y = np.array(df['success'])
y =  np.reshape(y,(y.shape[0],1))
X =  np.array(df.iloc[ : , 0:2])
X = np.hstack((np.ones((X.shape[0], 1)), X))
W = np.array([0.05 , 0.05 ,0.05]).reshape((3,1))
print(X.shape , y.shape ,W.shape)
```

```python
X1 = df[df['success']  == 0]
X2 = df[df['success']  == 1]
def LDA(X1,X2):
    X1_mean = np.array([X1["age"].mean(), X1['interest'].mean()])
    X2_mean = np.array([X2["age"].mean(), X2['interest'].mean()])  
    print(X1_mean , X2_mean)
    X1 = X1.iloc[: , 0:2]
    X2 = X2.iloc[: , 0:2]
    X1 =  X1 - X1.mean(axis=0)
    X2 =  X2 - X2.mean(axis=0)
    S1 = (X1.T@X1) / X1.shape[0]
    S2= (X2.T@X2) / X2.shape[0]
    SW = S1 + S2
    V = np.linalg.inv(SW) @ (X1_mean-X2_mean)
    return V
```

```python

def polyFeatureGen(df, deg):
    new_df =  pd.DataFrame()
    for i in range(deg+1):
        for j in range(deg+1):
            if i+j>deg:
                continue
            else:
                col_name = "x1^"+str(i)+'x2^'+str(j)
                new_df[col_name] = df['age']**i * df['interest']**j
    return new_df
```

```python
def mapFeature(X1, X2, degree):
    res = np.ones(X1.shape[0])
    for i in range(1,degree + 1):
        for j in range(0,i + 1):
            res = np.column_stack((res, (X1 ** (i-j)) * (X2 ** j)))
    
    return res

def plotDecisionBoundary(theta,degree, axes):
    u = np.linspace(-2.5, 2.5, 50)
    v = np.linspace(-2.5, 2.5, 50)
    U,V = np.meshgrid(u,v)

    U = np.ravel(U)
    V = np.ravel(V)
    Z = np.zeros((len(u) * len(v)))
    
    X_poly = mapFeature(U, V, degree)
    Z = X_poly.dot(theta)
    
   
    U = U.reshape((len(u), len(v)))
    V = V.reshape((len(u), len(v)))
    Z = Z.reshape((len(u), len(v)))
    
    cs = axes.contour(U,V,Z,levels=[0],cmap= "Greys_r")
    return cs

```

```python
df_input =  df.iloc[:, 0:2]
poly_features =  polyFeatureGen(df_input, 1)
for col in poly_features.columns:
    if col != 'x1^0x2^0':
        poly_features[col] = (poly_features[col] - poly_features[col].mean(axis=0))/poly_features[col].std(axis=0)
W_input = (np.ones((poly_features.shape[1],1)) * 0.1)
losses , W_poly = gradient_descent(poly_features,y,W_input,learning_rate=0.01, num_epochs=5000,verbose=0)

fig, axes = plt.subplots();
x_plot =  np.array(poly_features['x1^1x2^0'])
y_plot =  np.array(poly_features['x1^0x2^1'])
col = ['red' if x==1 else 'blue' for x in y]
axes.scatter(x =y_plot  , y=x_plot,c=col )
plotDecisionBoundary(W_poly, 1, axes)
```
